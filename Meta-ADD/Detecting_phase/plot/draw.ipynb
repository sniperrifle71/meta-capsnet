```python
import pandas as pd
import matplotlib.pyplot as plt


```

```python
# 数据
x = ['A', 'B', 'C', 'D']
y = [10, 7, 5, 3]

# 创建一个图形
fig, ax = plt.subplots()

# 绘制柱状图
ax.bar(x, y)

# 设置标签和标题
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Bar Chart')

# 显示图形
plt.show()
```

```python
df = pd.read_csv("./capsnet.csv")
df = df.transpose()
```

```python
df.head(10)
```

```python
dict_ = {}
for row in df.iterrows():
    dict_ = {}
    for sample in row[1].dropna():
        if sample not in dict_.keys():
            dict_[sample] = 1
        else:
            dict_[sample] += 1
    # 生成数据
    data = dict_
    print(row[0])
    # 绘制柱状图
    fig, ax = plt.subplots(figsize=(50, 6))
    ax.hist(data, bins=50, alpha=0.7, color='steelblue', edgecolor='none',width = 5)

    # 设置柱状图的参数
    ax.set_xlim(0, 3500)
    ax.set_ylabel('Frequency')
    ax.set_xlabel('Value')
    ax.set_title('Histogram of Random Data')

    # 调整子图参数
    plt.tight_layout()

    # 显示图形
    plt.show()

    
        
```
